```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_excel(r'Fig7-uc-individual&pixel-level.xlsx')
data.head()

plt.figure(figsize=(15, 15))
# Epis vs Dice
sns.lmplot(
    x='Epis',
    y='Dice',
    data=data
)
plt.show()
# Alea vs Dice
sns.lmplot(
    x='Alea',
    y='Dice',
    data=data
)
plt.show()

# Alea+Epis vs Dice
# norm to 0-1
data['Alea+Epis'] = data['Alea+Epis'] - data['Alea+Epis'].min()
data['Alea+Epis'] = data['Alea+Epis'] * 1.0 / data['Alea+Epis'].max()
sns.lmplot(
    x='Alea+Epis',
    y='Dice',
    data=data,

)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel(xlabel='Uncertainty', fontsize=17)
plt.ylabel(ylabel='Dice', fontsize=17)
plt.show()

# 2022/01/18 sqrt(Alea) + sqrt(Epis) vs Dice 
data['sqrt'] = np.sqrt(data['Alea']) + np.sqrt(data['Epis'])
# norm to 0-1
data['sqrt'] = data['sqrt'] - data['sqrt'].min()
data['sqrt'] = data['sqrt'] * 1.0 / data['sqrt'].max()
sns.lmplot(
    x='sqrt',
    y='Dice',
    data=data,

)
plt.show()  
```
